Исходная таблица

```python
import pandas as pd

df = pd.read_csv("titanic.csv")
df
```

1. Разбиения таблицы по полу

```python
df_female = df[df['Sex'] == 'female']
df_female
```

```python
df_male = df[df['Sex'] == 'male']
df_male
```

2. Удаление пустых строк

```python
df_cleaned = df.dropna()
df_cleaned
```

3. Построить гистограмму распределения пассажиров по возрасту

```python
import matplotlib.pyplot as plt

df['Age'].plot(kind='hist', bins=100, alpha=0.7)
plt.xlabel('Возраст')
plt.ylabel('Количество пассажиров')
plt.grid(axis='y')
```

4. Вычислить мин и макс, среднее, моду, медиану возраста пассажира каждого пола

```python
def calc(df):
    return {
        'min': df['Age'].min(),
        'max': df['Age'].max(),
        'mean': df['Age'].mean(),
        'mode': df['Age'].mode()[0],
        'median': df['Age'].median()
    }

stats_male = calc(df_male.dropna())
stats_female = calc(df_female.dropna())

pd.DataFrame({
    'Male': stats_male,
    'Female': stats_female
})
```

5. Визуализировать возраст при помощи различных настроек метода plot

```python
df_male['Age'].plot(kind='line', label='Male', color='blue', linestyle='-')
df_female['Age'].plot(kind='line', label='Female', color='pink', linestyle='-')

plt.title('Линейный график')
plt.ylabel('Возраст')
plt.legend()
plt.show()
```

```python
df_male['Age'].plot(kind='hist', alpha=0.5, label='Male', color='blue', bins=10)
df_female['Age'].plot(kind='hist', alpha=0.5, label='Female', color='pink', bins=10)

plt.title('Гистограмма')
plt.xlabel('Возраст')
plt.ylabel('Частота')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df_male['Age'].plot(kind='box', label='Male', color='blue')
df_female['Age'].plot(kind='box', label='Female', color='pink')

plt.title('Коробчатая диаграмма')
plt.ylabel('Возраст')
plt.show()
```

```python
df_male['Age'].plot(kind='kde', label='Male', color='blue', linestyle='--')
df_female['Age'].plot(kind='kde', label='Female', color='pink', linestyle='-.')

plt.title('График плотности')
plt.xlabel('Возраст')
plt.ylabel('Плотность')
plt.legend()
plt.show()
```

```python
plt.scatter(df_male.index, df_male['Age'], label='Male', color='blue', alpha=0.6)
plt.scatter(df_female.index, df_female['Age'], label='Female', color='pink', alpha=0.6)

plt.title('Точечная диаграмма')
plt.ylabel('Возраст')
plt.legend()
plt.show()
```
